```python
import pandas as pd       # pandas library brought in for data prep

df_ireland=pd.read_csv("ireland_production_index.csv")    # Introduced construction production index data for Ireland

df_ireland.head(5)
```

```python
df_ireland.info()     # All column data are floats except the year column, 93 rows and 5 columns
```

```python
df_ireland.isnull().sum()      # Second row is unnecessary and causing null values to appear
```

```python
df_ireland=df_ireland.drop(df_ireland.index[0])     # dropped the second row

df_ireland.head(5)
```

```python
df_ireland=df_ireland.reset_index(drop=True)   # corrected the index of the rows

df_ireland.head(5)
```

```python
df_ireland=df_ireland.rename(columns={'Unnamed: 0':'Time Period'})     # Assigned more appropriate name to first column

df_ireland.head(5)
```

```python
df_ireland.duplicated().sum()            # Validated no duplicate data
```

```python
df_ireland=df_ireland.rename(columns={'Volume of Production Index in Building and Construction (Seasonally Adjusted)':'Adjusted Volume'})

df_ireland.head(5)          # Final column has been renamed as will be used for merge and analysis
```

```python
columns_to_drop=df_ireland.columns[1:4]     #Removing unnecessary columns

df_ireland=df_ireland.drop(columns=columns_to_drop)

df_ireland.head(5)
```

```python
df_ireland['Year']=df_ireland['Time Period'].str[:4]

df_ireland['Average']=df_ireland.groupby('Year')['Adjusted Volume'].transform('mean')        # Got avg of qtrs within each year

df_ireland.head(5)
```

```python
df_ireland=df_ireland.drop_duplicates(subset='Average')           # Remove duplicate values for 'Average' column

df_ireland=df_ireland.drop(df_ireland.columns[[0,1]], axis=1) # Dropped first and second column


df_ireland.head(5)
```

```python
df_ireland=df_ireland.reset_index(drop=True)      # Reset row indices

df_ireland.head(5)
```

```python
df_ireland.describe() #EDA on ireland production dataset
```

```python
import matplotlib.pyplot as plt    # Library to visualise data
%matplotlib inline

df_ireland.boxplot()               # Data skews to the left

plt.ylabel('Normalised Production per year')
plt.title('Ireland Normalised Construction Production')
```

```python
df_denmark=pd.read_csv("denmark_production_index.csv")    # Introduce construction production index data from Denmark dataset

df_denmark.head(5)
```

```python
df_denmark=df_denmark.rename(columns={'DNKPROCONAISMEI':'Adjusted Volume'})

df_denmark.head(5)          # Second column has been renamed as will be used for merge and analysis
```

```python
df_denmark=df_denmark.drop(df_denmark.index[:5]) # Dropped the years which can't be compared to ireland dataframe


df_denmark.head(5)
```

```python
df_denmark=df_denmark.reset_index(drop=True)   #Reset the index

df_denmark.head(5)
```

```python
df_denmark['DATE']=df_denmark['DATE'].str[6:]  # Removed dd/mm/yyyy format to just yyyy to align with Ireland format


df_denmark.head(5)
```

```python
df_denmark.info()        # No need to check for nulls aswell since info function does
```

```python
df_denmark.boxplot()               

plt.ylabel('Production per year')
plt.title('Denmark Construction Production')   # Data skews slightly to the left but more normally distributed than Ireland
```

```python
selected_ireland_columns=df_ireland[['Year', 'Average']]
selected_denmark_columns=df_denmark[['DATE', 'Adjusted Volume']]

df_merged=pd.concat([selected_ireland_columns, selected_denmark_columns], axis=1)

df_merged.head(10)         #New dataframe with selected columns from the two previous dataframes
```

```python
df_merged=df_merged.drop(columns=['DATE'])    # Removed unnecessary DATE column

df_merged.head(5)
```

```python
df_merged=df_merged.rename(columns={'Average': 'Ireland', 'Adjusted Volume': 'Denmark'})   # Renamed columns for clarity


print(df_merged)
```

```python
x=df_merged['Year'] 
y1=df_merged['Ireland']
y2=df_merged['Denmark']

fig, ax1=plt.subplots()                # Plotting the trend between Ireland and Denmark for adjusted construction production
ax2=ax1.twinx()

ax1.plot(x, y1, label='Ireland', color='green')
ax2.plot(x, y2, label='Denmark', color='r')

ax1.set_xlabel('Year', fontsize=20)
ax1.set_ylabel('Ireland', fontsize=20)
ax2.set_ylabel('Denmark', fontsize=20)

legend1=ax1.legend(loc='upper left', fontsize=20)
legend2=ax2.legend(loc='upper center', fontsize=20)

fig.set_size_inches(14, 7)

ax1.set_title('Adjusted Construction Production', fontsize=30)

plt.show()
```
