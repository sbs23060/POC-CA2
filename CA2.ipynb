```python
from scipy.stats import norm                                          # Import norm module for calculating Z-testing

confidence_level=0.95                                                 # Assigning confidence level

sample_mean=df_merged['Ireland'].mean()                               # Taking sample mean
sample_std=df_merged['Ireland'].std()                                 

z_critical=norm.ppf((1+confidence_level)/2)                           # Calculating the critical value

margin_of_error=z_critical*(sample_std/(len(df_ireland)**0.5))        # Calculating the margin of error

lower_bound=sample_mean-margin_of_error                               # Calculating lower and upper bounds
upper_bound=sample_mean+margin_of_error

print("Ireland's lower bound is:", lower_bound)                       # Printing results
print("Ireland's upper bound is:", upper_bound)
print("Ireland's {}% confidence interval for the mean is [{:.2f}, {:.2f}]".format(int(confidence_level*100), lower_bound, upper_bound))
```
